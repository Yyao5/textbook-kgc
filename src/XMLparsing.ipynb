```python
index_table = {}
```

```python
def extract_pair(ref_list, table):
  seg_no = []
  if 'ref' in ref_list.keys():
    if ref_list['ref'] and type(ref_list['ref'])== dict:
      seg_no.append(ref_list['ref']['@target'])
    elif ref_list['ref'] and type(ref_list['ref'])== list:
      for x in ref_list['ref']:
        seg_no.append(x['@target'])
    term = ref_list['#text']
    for id in seg_no:
      if id not in table:
        table[id] = [term]
        print("creating new")
      elif term not in table[id]:
        table[id].append(term)
        print("adding new")
```

```python
for entry in doc['TEI']['back']['div']['list']['item']:
  if type(entry) == dict:
    print("dict start")
    extract_pair(entry, index_table)
    # elif type(entry) == 'list':
    #   for sub in entry:
    #     extract_pair(entry)
    if len(entry) > 2:
      sub_entry = entry['list']['item']

      if type(sub_entry) == dict:
              extract_pair(sub, index_table)
      elif type(sub_entry) == list:
            for sub in sub_entry:
              extract_pair(sub, index_table)
              if len(sub) > 2:
                sub_sub_entry = sub['list']['item']

                if type(sub_sub_entry) == list:
                  for sub_sub in sub_sub_entry:
                    extract_pair(sub_sub, index_table)
                elif type(sub_sub_entry) == dict:
                    extract_pair(sub_sub, index_table)
```

```python
def extract_title(numbered_title):
  # if len(numbered_title.split()) > 1:
  #     title_seg = numbered_title.split()[alpha_start_ind:]
  #     title_only = ' '.join(title_seg)
  # else:
  #     title_only = numbered_title
  if ' '.join(numbered_title.split()[1:]) == 'exercises and problems':
    return None
  return numbered_title
```

```python
content_table = {}
for obj in doc['TEI']['front']['div']['list']['item']:
  # print(obj)
  seg = obj['ref']['@target']
  chap = extract_title(obj['#text'])
  if chap:
    content_table[chap] = seg
```

```python
for obj in doc['TEI']['front']['div']['list']['list']:
  # print(obj[1]['item'])
  for subsec in obj['item']:
    # print(subsec[1]['#text'].split()[1:])
    seg = subsec['ref']['@target']
    subchap = extract_title(subsec['#text'])
    # if len(subsec['#text'].split()) > 1:
    #   chap_seg = subsec['#text'].split()[1:]
    #   chap = ' '.join(chap_seg)
    # else:
    #   chap = subsec['#text']
    if seg not in content_table and subchap:
      content_table[subchap] = seg
```

```python
content_table_kg = []
for i in range(len(doc['TEI']['front']['div']['list']['list'])):
  subsections = doc['TEI']['front']['div']['list']['list'][i]['item']
  for subsection in subsections:
    subject = extract_title(doc['TEI']['front']['div']['list']['item'][i]['#text'])
    predicate = "contains"
    obj =  extract_title(subsection['#text'])
    if subject and obj:
      content_table_kg.append([subject, predicate, obj])
```

```python
for title in content_table:
  if content_table[title] in index_table:
    for term in index_table[content_table[title]]:
      content_table_kg.append([title, "contains", term])
```
